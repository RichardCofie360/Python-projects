```python
import pandas as pd

# Sample DataFrame
data = {'Sales': [100, 200, 300, 150, 250],
        'Products': [5, 10, 15, 8, 12],
        'State': ['NY', 'CA', 'TX', 'NY', 'CA']}

df = pd.DataFrame(data)

# Group by 'State' and calculate the sum of 'Sales' and 'Products'
result_df = df.groupby('State').agg({'Sales': 'sum', 'Products': 'sum'}).reset_index()

# Display the result DataFrame
print(result_df)
```
